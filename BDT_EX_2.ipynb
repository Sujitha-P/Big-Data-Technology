```python
import os
import shutil

# Simulated HDFS directory
HDFS_DIR = "/content/hdfs"

# Create HDFS directory (if not exists)
os.makedirs(HDFS_DIR, exist_ok=True)

print("HDFS simulation folder created at:", HDFS_DIR)
```

```python
# Create a sample local file
local_file = "/content/local_sample.txt"
with open(local_file, "w") as f:
    f.write("This is a sample file to upload to HDFS.\n")

# Upload to HDFS
hdfs_file = os.path.join(HDFS_DIR, "uploaded_sample.txt")
shutil.copy(local_file, hdfs_file)

print("File uploaded to HDFS:", hdfs_file)
```

```python
download_path = "/content/downloaded_sample.txt"
shutil.copy(hdfs_file, download_path)

print("File downloaded from HDFS to local:", download_path)
```

```python
hdfs_copy = os.path.join(HDFS_DIR, "copied_sample.txt")
shutil.copy(hdfs_file, hdfs_copy)

print("File copied inside HDFS:", hdfs_copy)
```

```python
hdfs_moved = os.path.join(HDFS_DIR, "moved_sample.txt")
shutil.move(hdfs_copy, hdfs_moved)

print("File moved inside HDFS:", hdfs_moved)
```

```python
# Remove the moved file
os.remove(hdfs_moved)
print("Removed file:", hdfs_moved)

```
